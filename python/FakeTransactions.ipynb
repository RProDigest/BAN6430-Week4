```python
import pandas as pd
import numpy as np
from faker import Faker
import random

```

```python
# Initialize faker instance
fake = Faker()

def dummyData(numRec):
    transDate = []
    custName = []
    cardNum = []
    zipCode = []
    transAmount = []
    transType = []  # For transaction types
    transChannel = []  # For transaction channels

    transaction_types = ['Debit Card', 'Online Transfer', 'ATM Withdrawal']
    channels = ['Branch', 'Mobile App', 'ATM', 'Website']

    for _ in range(numRec):
        transDate.append(fake.date_time_between(start_date='-5y', end_date='now'))
        custName.append(fake.name())
        cardNum.append(fake.credit_card_number(card_type=None))  # Generating a random credit card number
        zipCode.append(fake.zipcode())
        transAmount.append(np.random.randint(100, 10000))  # Random transaction amount between 100 and 10,000
        transType.append(random.choice(transaction_types))  # Random transaction type
        transChannel.append(random.choice(channels))  # Random transaction channel

    # Creating a DataFrame
    dfDummyTrans = pd.DataFrame(list(zip(transDate, custName, cardNum, zipCode, transAmount, transType, transChannel)),
                                columns=['Transaction Date', 'Customer Name', 'Card Number', 'Zip Code', 
                                         'Transaction Amount', 'Transaction Type', 'Transaction Channel'])

    return dfDummyTrans

# Generating 50000 fake transactional records
dfDummyTrans = dummyData(50000)

# Displaying the data
dfDummyTrans.head()
```
